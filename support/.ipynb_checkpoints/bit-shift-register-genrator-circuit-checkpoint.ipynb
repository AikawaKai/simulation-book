```python
%matplotlib notebook

import SchemDraw as schem
import SchemDraw.elements as e
import SchemDraw.logic as l

d = schem.Drawing()
xor = d.add(l.XOR2)
d.add(e.RBOX, botlabel='$b_0$')
d.add(e.DOT)
d.push()
d.add(e.ARROWLINE, d='down', toy=d.unit*0.5)
d.pop()
d.add(e.RBOX, botlabel='$b_1$')
d.add(e.DOT)
d.push()
d.add(e.ARROWLINE, d='down', toy=d.unit*0.5)
d.pop()
d.push()
d.add(e.RBOX, botlabel='$b_2$')
d.add(e.DOT)
d.push()
d.add(e.ARROWLINE, d='down', toy=d.unit*0.5)
d.pop()
d.add(e.LINE, d='up', l=d.unit*.4)
d.add(e.LINE, d='left', tox=xor.in2-.3)
d.add(e.LINE, d='down', toy=xor.in2)
d.add(e.LINE, d='right', tox=xor.in2)
d.pop()
d.add(e.LINE, d='up', l=d.unit*.3)
d.add(e.LINE, d='left', tox=xor.in1)
d.add(e.LINE, d='down', toy=xor.in1)
d.draw()
```
